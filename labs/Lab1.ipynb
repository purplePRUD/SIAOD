### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
user_m = int(input())
user_n = int(input())
user_min_limit = int(input())
user_max_limit = int(input())

def PrintMatrix (arr):
    
    for row in arr:
        for element in row:
            print("\t" + str(element), end = " ")
        print()
        

def randMatrix (user_m, user_n, user_min_limit, user_max_limit):
    
    Matrix = []
    
    for i in range(user_m):
        Row = []
        for j in range(user_n):
            Row.append(random.randint(user_min_limit, user_max_limit))   
        Matrix.append(Row)
   
    return Matrix

matrix = randMatrix(user_m, user_n, user_min_limit, user_max_limit)
            
PrintMatrix(matrix)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
import time
import copy
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()

first = copy.copy(matrix)

for row in range(len(first)):
    number = len(first[row])

    while number > 0:
        maxEll = first[row][0]
        index = 0
        for i in range(number):
            if maxEll < first[row][i]:
                maxEll = first[row][i]
                index = i
            if i == number-1:
                mem = first[row][number-1]
                first[row][number-1] = maxEll
                first[row][index] = mem
        number -= 1
        
choiceSortTime = round((time.time() - start_time)*1000)

print()
PrintMatrix(first)

#TODO
print("--- {0} ms ---".format(choiceSortTime))
```

```python
# Сортировка вставкой.
start_time = time.time()

second = copy.copy(matrix)

for row in range(len(second)):
    number = len(second[row])

    for i in range(1, number):
        for j in range(i, 0, -1):
            if second[row][j] < second[row][j-1]:
                second[row][j], second[row][j-1] = second[row][j-1], second[row][j]
            else:
                break
                
insertSortTime = round((time.time() - start_time)*1000)

PrintMatrix(second)

#TODO
print("--- {0} ms ---".format(insertSortTime))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()

third = copy.copy(matrix)

for row in range(len(third)):
    number = len(third[row])

    for i in range(number-1):
        for j in range(number-1-i):
            if third[row][j] > third[row][j+1]:
                third[row][j], third[row][j+1] = third[row][j+1], third[row][j]
                
changeSortTime = round((time.time() - start_time)*1000)

print()
PrintMatrix(third)

#TODO
print("--- {0} ms ---".format(changeSortTime))
```

```python
# Сортировка Шелла.
start_time = time.time()

fourth = copy.copy(matrix)

for row in range(len(fourth)):
    number = len(fourth[row])
    diapason = number // 2

    while diapason > 0:
        for value in range(diapason, number):
            currentValue = fourth[row][value]
            index = value

            while index >= diapason and fourth[row][index - diapason] > currentValue:
                fourth[row][index] = fourth[row][index - diapason]
                index -= diapason
                fourth[row][index] = currentValue

        diapason //= 2
        
shellSortTime = round((time.time() - start_time)*1000)

print()
PrintMatrix(fourth)

#TODO
print("--- {0} ms ---".format(shellSortTime))
```

```python
# Быстрая сортировка.

def quickSort(arr):
    if len(arr) <= 1:
        return arr

    startElement = arr[0]
    lowArr = list(filter(lambda x: x < startElement, arr))
    midArr = [i for i in arr if i == startElement]
    highArr = list(filter(lambda x: x > startElement, arr))

    return quickSort(lowArr) + midArr + quickSort(highArr)


start_time = time.time()

fifth = copy.copy(matrix)

for row in range(len(fifth)):
    fifth[row] = quickSort(fifth[row])
    
quickSortTime = round((time.time() - start_time)*1000)

print()
PrintMatrix(fifth)

#TODO
print("--- {0} ms ---".format(quickSortTime))
    
```

```python
# Турнирная сортировка.
start_time = time.time()

def heapSort(arr):
    maxHeap(arr)
    for i in range(len(arr) - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, index=0, size=i)


def parent(i):
    return (i - 1) // 2


def left(i):
    return 2 * i + 1


def right(i):
    return 2 * i + 2


def maxHeap(arr):
    length = len(arr)
    start = parent(length - 1)
    while start >= 0:
        heapify(arr, index=start, size=length)
        start = start - 1


def heapify(arr, index, size):
    l = left(index)
    r = right(index)
    if (l < size and arr[l] > arr[index]):
        largest = l
    else:
        largest = index
    if (r < size and arr[r] > arr[largest]):
        largest = r
    if (largest != index):
        arr[largest], arr[index] = arr[index], arr[largest]
        heapify(arr, largest, size)


sixth = copy.copy(matrix)

for row in range(len(sixth)):
    heapSort(sixth[row])
    
tournamentSortTime = round((time.time() - start_time)*1000)

PrintMatrix(sixth)

#TODO
print("--- {0} ms ---".format(tournamentSortTime))
```

### Вывод


```python
start_time = time.time()

standart = copy.copy(matrix)

for i in range(len(standart)):
    standart[i].sort()

standartSortTime = round((time.time() - start_time)*1000)

times = [[choiceSortTime,"choice sort:"],
           [insertSortTime, "insert sort:"],
           [changeSortTime, "change sort:"],
           [shellSortTime, "shell sort: "],
           [quickSortTime, "quick sort: "],
           [tournamentSortTime, "tournament sort:"],
           [standartSortTime, "standart sort:"]]
times = sorted(times, key= lambda times: times[0])

for i in range(len(times)):
    print(str(i+1) + ") ", times[i][1], "\t", times[i][0], "ms")
```
